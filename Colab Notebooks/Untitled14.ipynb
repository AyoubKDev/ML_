```python
blood_columns = df.columns[(missing_rate < 0.9) & (missing_rate >0.88)]
viral_columns = df.columns[(missing_rate < 0.88) & (missing_rate >0.75)]
for col in blood_columns:
  plt.figure()
sns.distplot(positive_df[col],label='1')
sns.distplot(negative_df[col], label='0')
plt.legend()

sns.countplot(x='Patient age quantile', hue='SARS-Cov-2 exam result', data=df)

pd.crosstab(df['SARS-Cov-2 exam result'], df['Influenza A'])

# Replace distplot with displot or histplot
for col in viral_columns:
    plt.figure()
    sns.histplot(positive_df[col], label='positive', kde=True)
    sns.histplot(negative_df[col], label='negative', kde=True)
    plt.legend()
    plt.show()
```

```python
# Update crosstab without annot parameter
for col in viral_columns:
    plt.figure()
    crosstab_result = pd.crosstab(df['SARS-Cov-2 exam result'], df[col])
    sns.heatmap(crosstab_result, annot=True, fmt='d', cmap='YlGnBu', cbar=False)
    plt.show()
```
