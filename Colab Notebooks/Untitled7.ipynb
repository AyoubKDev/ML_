```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/ratings_Beauty.csv")
print("Columns:", dataset.columns)
print("Number of Rows:", len(dataset))
statistics = dataset.describe()
print("Averaged Numbers, Max/Min Values:\n", statistics)
missing_values = dataset.isna().sum()
print("Number of Missing Values per Column:\n", missing_values)
```

```python
dataset.sort_values(by=['ProductId'], ascending=False)
```

```python
grouper = pd.Grouper(key='Timestamp', freq='D')

# Apply the Grouper object to the dataset
grouped_dataset = dataset.groupby('Timestamp')

# Extract the size attribute of the Grouper object, which contains the number of rows in each group
days = grouped_dataset.size()

# Add a new column to the dataset containing the days
dataset['days'] = days
dataset
# Save the updated dataset
#df.to_csv('updated_dataset.csv', index=False)
```
