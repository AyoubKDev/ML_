```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
feature1 = [4, 1, 9, 7, 10, 12, 14, 8]
feature2 = [1, 2, 14, 5, 16, 18, 10, 10]

# Visualize the data using a scatter plot
plt.scatter(feature1, feature2, label='Data Points')
plt.scatter(11, 3, color='red', marker='x', label='Test Point (11, 3)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Data Points')
plt.legend()
plt.show()
```

```python
from math import sqrt

# Test point
test_point = np.array([11, 3])

# Training data
training_data = np.array(list(zip(feature1, feature2)))

# Calculate Euclidean distance from all training data points
distances = [sqrt(np.sum((test_point - x) ** 2)) for x in training_data]

# Sort the distances and get the indices of the first k neighbors
k = 4
nearest_neighbors_indices = np.argsort(distances)[:k]

# Print the indices of the k nearest neighbors
print("Indices of the k nearest neighbors:", nearest_neighbors_indices)
```

```python
from sklearn.neighbors import NearestNeighbors

# Reshape the training data for NearestNeighbors
training_data_reshaped = np.array(list(zip(feature1, feature2)))

# Create a NearestNeighbors model
knn_model = NearestNeighbors(n_neighbors=k)
knn_model.fit(training_data_reshaped)

# Find the k nearest neighbors for the test point using the kneighbors method
distances, indices = knn_model.kneighbors([test_point])

# Print the indices of the k nearest neighbors
print("Indices of the k nearest neighbors using sklearn:", indices[0])
```

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
print(iris)
# Create and train a KNeighborsClassifier model
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X, y)

# Given observations
observations = np.array([[5.4, 3.3, 5.9, 1.1], [1, 1, 1, 1]])

# Predict their classes based on the class of their neighbors
predictions = knn_model.predict(observations)

# Print the predicted classes
print("Predicted classes:", predictions)
```
