```python
import pandas as pd

df = pd.read_csv('/content/tested.csv')

#print df to check
print(df)
```

```python
selected_columns = ["Survived", "Age", "Fare","Sex","Pclass"]
selected_df = df[selected_columns]
selected_df
```

```python
df.dropna(inplace=True)
df
```

```python
# Convert "Sex" column to a categorical type
df["Sex"] = pd.Categorical(df["Sex"])

# Assign category codes (0 for female, 1 for male)
df["Sex"] = df["Sex"].cat.codes
df
```

```python
# Separate features and target
X = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Ticket", "Fare", "Cabin", "Embarked"]]
y = df["Survived"]

# Check the dimensions
print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.neighbors import KNeighborsClassifier


# Stratified sampling to ensure balanced class distribution in splits
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=1)

# Get training and testing indices
train_index, test_index = next(iter(sss.split(X, y)))

# Split data using indices
X_train, X_test = X.iloc[train_index], X.iloc[test_index]
y_train, y_test = y.iloc[train_index], y.iloc[test_index]

# Create a KNeighborsClassifier with distance metric optimization
knn = KNeighborsClassifier(metric='manhattan')
knn
# Train the model
knn.fit(X_train, y_train)

# Make predictions
y_pred = knn.predict(X_test)

# Evaluate the model (e.g., accuracy)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")


```
