```python
import pandas as pd
import numpy as np

# Load the dataset with unnamed columns
dataset = pd.read_csv('/content/oil-spill.csv', header=None)

# Rename the columns as col1, col2, ...
dataset.columns = [f'col{i}' for i in range(1, len(dataset.columns) + 1)]

# 3.1 Visual analysis of the dataset
print("Dataset Preview:")
print(dataset.head())

# 3.2 Identify columns with a single value
single_value_columns = dataset.columns[dataset.nunique() == 1]
print("\nColumns with a single value:")
print(single_value_columns)

# 3.3 function to summarize the number of unique values for each column
def summarize_unique_values(data):
    unique_counts = data.nunique()
    return unique_counts

# 4. Remove columns containing a single value
def remove_single_value_columns(data):
    single_value_columns = data.columns[data.nunique() == 1]
    data = data.drop(columns=single_value_columns)
    return data

# 5. Consider columns with very few unique values
# 5.1 Write code to calculate the percentage of unique values
def calculate_percentage_unique_values(data):
    total_rows = len(data)
    percentage_unique_values = (data.nunique() / total_rows) * 100
    filtered_percentages = percentage_unique_values[percentage_unique_values > 1]
    return filtered_percentages

# Example Usage:
# Summary of unique values for each column
summary_unique_values = summarize_unique_values(dataset)
print("\nSummary of unique values for each column:")
print(summary_unique_values)

# Remove columns with a single value
dataset_removed_single_value_columns = remove_single_value_columns(dataset)
print("\nDataset after removing columns with a single value:")
print(dataset_removed_single_value_columns)

# Calculate the percentage of unique values for each column
percentage_unique_values = calculate_percentage_unique_values(dataset)
print("\nPercentage of unique values for each column:")
print(percentage_unique_values)
```
