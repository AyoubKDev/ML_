```python
# Importer les bibliothèques nécessaires
import pandas as pd
from sklearn.datasets import load_iris
from io import StringIO

# Question 1 : Téléchargez l'ensemble de données et affichez-en une partie
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(data.head(-10))


```

```python
# Question 2 : Écrire les données au format CSV
csv_data = StringIO()
data.to_csv(csv_data, index=False)

# Afficher le contenu CSV
print("Contenu CSV:")
print(csv_data.getvalue())

```

```python
# Question 3 : Écrire des données au format JSON
json_data = StringIO()
data.to_json(json_data, orient='records')

# Afficher le contenu JSON
print("Contenu JSON:")
print(json_data.getvalue())
```

```python
import xml.etree.ElementTree as ET
import json

# Chargement du fichier XML
xml_file_path = '/content/file.xml'
tree = ET.parse(xml_file_path)
root = tree.getroot()

# Fonction récursive pour convertir l'élément XML en dictionnaire
def xml_to_dict(element):
    result = {}
    for child in element:
        child_data = xml_to_dict(child)
        if child_data:
            result[child.tag] = child_data
        else:
            result[child.tag] = child.text
    return result

# Convertir l'élément racine en dictionnaire
json_data = xml_to_dict(root)

# Convertir le dictionnaire en format JSON
json_string = json.dumps(json_data, indent=2)

# Afficher le code JSON résultant
print(json_string)

# Sauvegarder le code JSON dans un fichier
json_output_file_path = '/content/fileOutPut.xml'  # Remplacez par le chemin de sortie souhaité
with open(json_output_file_path, 'w') as json_file:
    json_file.write(json_string)
```
