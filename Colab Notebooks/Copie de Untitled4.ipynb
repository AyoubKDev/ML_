```python
import pandas as pd
dataset = pd.read_csv('./train.csv')

print("Columns:", dataset.columns)
print("Number of Rows:", len(dataset))
statistics = dataset.describe()
print("Averaged Numbers, Max/Min Values:\n", statistics)
missing_values = dataset.isna().sum()
print("Number of Missing Values per Column:\n", missing_values)


# Display the first 10 rows of the dataset
first_10_rows = dataset.head(10)
print("First 10 Rows of the Dataset:\n", first_10_rows)

# Number of unique values per column
unique_values = dataset.nunique()

# Data types of values in each column
data_types = dataset.dtypes

print("Number of Unique Values per Column:\n", unique_values)
print("Data Types of Values in Each Column:\n", data_types)

# Group the dataset by a specific column (e.g., "order date") and calculate averages
grouped_data = dataset.groupby('order date').mean()

# Find the row with the highest average
row_with_highest_average = grouped_data.mean().idxmax()

# Define the desired order of columns
desired_columns = ['important_column1', 'important_column2', 'important_column3', 'other_column1', 'other_column2']

# Create a new DataFrame with columns in the desired order
rearranged_dataset = dataset[desired_columns]
```
