```python
class ConcourProblem(Problem):
  def __init__(self):
    self.goal=(10,2,1,6)
    self.initial = tuple([-1]*4)
    super().__init__((-1,-1,-1,-1),self.goal)
    


  def actions(self,state):
    if state[-1] != -1 :
      return []
    else:
      return[num for num in range(1,11) 
      if num not in state]
  
  def result(self, state, num):
    index = state.index(-1)
    new = list(state[:])
    new[index] = num
    return tuple(new)
```

```python
Problem=ConcourProblem()
solution = depth_first_tree_search(Problem)
print(solution)
print(solution.path)
```
